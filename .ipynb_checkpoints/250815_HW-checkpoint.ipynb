```python
import pandas as pd
import numpy as np

df = pd.read_csv("prices.csv")

df["Date"] = pd.to_datetime(df["Date"], utc=True)
```

```python
df["Date"] = df["Date"].dt.tz_localize(None)
```

```python
df_filtered = df[
    (df["Ticker"] == "NVDA") &
    (df["Country"] == "usa") &
    (df["Date"].dt.year.between(2000, 2005))
]
df_filtered
```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(4, 1, figsize=(14, 12), sharex=True)

axes[0].plot(df_filtered["Date"], df_filtered["Open"], label="Open")
axes[0].plot(df_filtered["Date"], df_filtered["High"], label="High")
axes[0].plot(df_filtered["Date"], df_filtered["Low"], label="Low")
axes[0].plot(df_filtered["Date"], df_filtered["Close"], label="Close")
axes[0].set_ylabel("Price")
axes[0].set_title("NVDA OHLC Prices (2000â€“2005)")
axes[0].legend()
axes[0].grid(True)

axes[1].plot(df_filtered["Date"], df_filtered["Volume"], color="purple")
axes[1].set_ylabel("Volume")
axes[1].set_title("NVDA Volume (2000â€“2005)")
axes[1].grid(True)

axes[2].plot(df_filtered["Date"], df_filtered["Stock Splits"], color="red")
axes[2].set_ylabel("Stock Splits")
axes[2].set_title("NVDA Stock Splits (2000â€“2005)")
axes[2].grid(True)

axes[3].plot(df_filtered["Date"], df_filtered["Dividends"], color="blue")
axes[3].set_ylabel("Dividends")
axes[3].set_title("NVDA Dividends (2000â€“2005)")
axes[3].grid(True)

plt.xlabel("Date")
plt.tight_layout()
plt.show()
```
